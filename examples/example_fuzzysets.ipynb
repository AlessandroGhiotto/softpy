# Discrete Fuzzy Sets

```python
oranges = DiscreteFuzzySet(["apple", "lemon", "oxanthera","orange"], [0.0, 0.2, 0.5, 1.0])
```

```python
oranges[0.2]
```

```python
try:
    oranges[-0.1]
except ValueError as v:
    print(v)
```

```python
plt.bar(oranges.items, oranges.memberships)
```

```python
try:
    print(oranges("pear"))
except ValueError as v:
    print(v)
```

```python
oranges2 = DiscreteFuzzySet(["apple", "lemon", "oxanthera","orange"], [0.0, 0.2, 0.5, 1.0])

oranges == oranges2
```

```python
oranges_nodynamic = DiscreteFuzzySet(["apple", "lemon", "oxanthera","orange"], [0.0, 0.2, 0.5, 1.0], dynamic=False)

oranges == oranges_nodynamic
```

```python
lemons = DiscreteFuzzySet(["apple", "lemon", "oxanthera", "orange", "bergamot"], [0.0, 1.0, 0.2, 0.3, 0.45])

lemons == oranges
```

```python
oranges3 = DiscreteFuzzySet(["apple", "lemon", "oxanthera","orange"], [0.0, 0.2, 0.3, 1.0])

oranges == oranges3
```

```python
oranges.hartley()
```

# Operations (Discrete)

```python
max_t = maximum(oranges, lemons)
max_t("orange")
```

```python
max_t[0.5]
```

```python
plt.bar(max_t.items, max_t.memberships)
```

```python
min_t = minimum(oranges, lemons)
min_t("orange")
```

```python
min_t[0.2]
```

```python
plt.bar(min_t.items, min_t.memberships)
plt.ylim(0,1)
```

```python
prod_t = product(oranges, lemons)
prod_t('orange')
```

```python
prod_t[0.2]
```

```python
plt.bar(prod_t.items, prod_t.memberships)
plt.ylim(0,1)
```

```python
owa_ex = owa([oranges, oranges_nodynamic, lemons], w=[0.25, 0.5, 0.25])
owa_ex('orange')
```

```python
owa_ex[0.2]
```

```python
plt.bar(owa_ex.items, owa_ex.memberships)
plt.ylim(0,1)
```

# Continuous Fuzzy Sets

```python
f = TriangularFuzzyNumber(0, 5, 10)
```

```python
xs = np.linspace(0, 10, 100)
ys = [f(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(0,10)
plt.ylim(0,1)
```

```python
low, upp = f[0.5]
xs = np.linspace(low, upp, 100)
ys = [f(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(0,10)
plt.ylim(0,1)
```

```python
vals = f[0.5]
print(vals)
plt.plot(vals, np.ones(len(vals)))
plt.xlim(0,10)
plt.ylim(0,1.1)
```

```python
f.fuzziness()
```

```python
f.hartley()
```

# Operations (Continuous)

```python
g = RampFuzzyNumber(5, 0)
xs = np.linspace(-1, 20, 100)
ys = [g(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(-1,20)
plt.ylim(0,1.05)
```

```python
hmax = maximum(f,g)
```

```python
xs = np.linspace(-1, 20, 100)
ys = [hmax(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(-1,20)
plt.ylim(0,1.05)
```

```python
vals = hmax[0.7]
plt.plot(vals, np.ones(len(vals)))
plt.xlim(0,20)
plt.ylim(0,1.1)
```

```python
hmin = minimum(f,g)
```

```python
xs = np.linspace(-1, 20, 100)
ys = [hmin(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(-1,20)
plt.ylim(0,1)
```

```python
vals = hmin[0.495]
plt.plot(vals, np.ones(len(vals)))
plt.xlim(0,20)
plt.ylim(0,1.1)
```

```python
h = TrapezoidalFuzzyNumber(5, 10, 15, 20)
xs = np.linspace(0, 30, 100)
ys = [h(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(0,25)
plt.ylim(0,1.05)
```

```python
owa_c = owa([f,g,h], [0.05, 0.05, 0.9])
```

```python
xs = np.linspace(-1, 20, 100)
ys = [owa_c(v) for v in xs]
plt.plot(xs, ys)
plt.xlim(-1,25)
plt.ylim(0,1)
```

```python
vals = owa_c[0.75]

plt.plot(vals, np.ones(len(vals)))
plt.xlim(-1,25)
plt.ylim(0,1.1)
```
